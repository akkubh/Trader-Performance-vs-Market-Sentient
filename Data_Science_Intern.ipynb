**Project Summary**

**Methodology**

This project analyzes the relationship between Bitcoin market sentiment and Hyperliquid trader performance by combining the Bitcoin Fear & Greed Index with historical trading data.


The workflow included:

*  Data loading and cleaning
*   Handling missing values and duplicate records
*   Timestamp conversion and daily-level data alignment
*  Feature engineering (daily PnL, trade frequency, profitability indicators, trader segmentation)
*   Exploratory Data Analysis (EDA)
*  Statistical comparison of trader performance across sentiment categories
*   Optional predictive modeling using a Random Forest Classifier



**Key Insights**

* Trader profitability varied across market sentiment categories, with Extreme Greed producing the highest average Closed PnL.
* Although average profits were positive, the median Closed PnL remained zero across all sentiment categories, indicating that a relatively small number of highly profitable trades contributed significantly to overall returns.
* Trading behavior changed with market sentiment, as reflected by differences in trade frequency, position sizes, and directional bias.
* Trader segmentation highlighted distinct behavioral patterns among frequent traders, profitable traders, and high-volume traders.


**Strategy Recommendations**

**Strategy 1 – Dynamic Risk Management**

During periods of Extreme Fear and Greed, traders should reduce position sizes and adopt stricter stop-loss strategies due to increased variability in trading outcomes.


**Strategy 2 – Sentiment-Aware Trade Selection**

Rather than increasing trading frequency, traders should prioritize high-confidence setups during favorable market conditions, as profitability appears to depend more on trade quality than trade quantity.

**Bonus**

A Random Forest Classifier was developed to predict trade profitability using trader behavior and market sentiment features.

The model achieved an accuracy of 98.83% on the test dataset. Feature importance analysis identified trading direction, start position, execution price, and trade side as the strongest predictors of profitability, demonstrating the usefulness of behavioral features in predictive trading models.

**Conclusion**

The analysis demonstrates that market sentiment influences trader behavior and performance. Integrating sentiment indicators with historical trading data provides valuable insights into trading patterns and supports the development of more informed, data-driven trading strategies.
